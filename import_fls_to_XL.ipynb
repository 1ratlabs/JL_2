```python
# This function opens a file that contains A # of readings from the ib and creates excel data sheet containing One row for each test
#  The "def read_csv_with_fallback" It is called by the“ process in body function”
# The"def process_inbody(folder_name)" Is called and external call statement that starts the process of importing files into the single worksheet.
# The Excel worksheet is stored in working directory
# The path to the directory of the named file in the Onedrive 
#  Path(r"/mnt/c/Users/bhuns/OneDrive/___Health Data/__DD studies/InBody CSV")
#=============================================================================
import pandas as pd 
from pathlib import Path 

def read_csv_with_fallback(path):
    """Try UTF‑8 first, then fallback encodings."""
    encodings_to_try = ["utf-8", "cp1252", "latin1"]

    for enc in encodings_to_try:
        try:
            return pd.read_csv(path, encoding=enc)
        except UnicodeDecodeError:
            continue

    raise UnicodeDecodeError(f"Could not decode file: {path}")


def process_inbody(folder_name):
    """
    Reads all CSV files inside the given folder and writes a combined Excel file
    named '{folder_name}_xl_ws.xlsx'. Warns if the output file is open in Excel.
    """

    base_path = Path(r"/mnt/c/Users/bhuns/OneDrive/___Health Data/__DD studies/InBody CSV")
    folder = base_path / folder_name

    csv_files = sorted(folder.glob("*.csv"))

    if not csv_files:
        raise FileNotFoundError(f"No CSV files found in folder '{folder_name}'")

    # Read with fallback encodings
    df_list = [read_csv_with_fallback(f) for f in csv_files]
    combined = pd.concat(df_list, ignore_index=True)

    # output_path = base_path / f"{folder_name}_xl_ws.xlsx" # onedrive  directory 
    output_path = Path.cwd() / f"{folder_name}_xl_ws.xlsx"

    print(output_path)

    # Try writing the file, catch "file in use" errors
    try:
        combined.to_excel(output_path, index=False)
        print(f"Created Excel file: {output_path}")

    except PermissionError:
        print(f"⚠️ The file '{output_path.name}' is currently open in Excel.")
        print("   Please close the file and run the function again.")
```

```python
# This function reads Excel file that contains the data from all tests; one row per test and builds the dataframe "ib_xl_ws_df(name)" [def build_ib_xl_ws_df(name):]
import pandas as pd
from pathlib import Path

def build_ib_xl_ws_df(name):
    """
    Reads an Excel file named '{name}.xlsx' from the working directory
    and returns a DataFrame named df_{name}raw.
    """

    # Build the filename
    file_path = Path(f"{name}_xl_ws.xlsx")

    # Check if file exists
    if not file_path.exists():
        raise FileNotFoundError(f"File '{file_path}' not found in working directory.")

    # Read Excel file
    df = pd.read_excel(file_path)

    # Create a dynamic variable name (optional)
    globals()[f"df_{name}raw"] = df

    print(f"Loaded Excel file: {file_path}")
    print(f"Created DataFrame: df_{name}raw")

    return df
```
