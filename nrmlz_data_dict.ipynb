# creating normalizing def functions

```python
def normalize_dict_to_mean_one(data_dict, timestamp_col="Cleaned_Timestamp"):
    """
    Normalizes each numeric column in a dict so its mean becomes 1, then subtracts 1.
    Skips the timestamp column.

    Parameters:
    - data_dict: dict — source dictionary with column names as keys
    - timestamp_col: str — name of the timestamp column to exclude

    Returns:
    - normalized_dict: dict — new dictionary with normalized values
    """
    import numpy as np
    import pandas as pd

    df = pd.DataFrame(data_dict)
    normalized_dict = {}

    for col in df.columns:
        if col == timestamp_col:
            normalized_dict[col] = df[col]  # preserve timestamp as-is
            continue

        # Drop NaNs before computing mean
        col_values = pd.to_numeric(df[col], errors="coerce")
        mean_val = col_values.mean()

        if pd.isna(mean_val) or mean_val == 0:
            normalized_dict[col] = [np.nan] * len(col_values)  # fallback if mean is invalid
        else:
            normalized_dict[col] = ((col_values / mean_val) - 1).tolist()

    return normalized_dict
```

## Calling functions to compute normalizations


```python
ib970mrn_grp1_nrml = normalize_dict_to_mean_one(ib970mrn_grp1)
```
